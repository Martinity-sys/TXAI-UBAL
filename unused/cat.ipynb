```python
import matplotlib.pyplot as plt

def draw_cat_with_lines():
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.axis('off')  # Hide axes

    # Head (circle-like)
    head = plt.Circle((0, 0), 1, fill=False, linewidth=2)
    ax.add_patch(head)

    # Ears (triangles)
    ax.plot([-0.6, -0.9, -0.4], [0.8, 1.4, 1.2], color='black')  # Left ear
    ax.plot([0.6, 0.9, 0.4], [0.8, 1.4, 1.2], color='black')     # Right ear

    # Eyes
    ax.plot([-0.4, -0.4], [0.3, 0.4], color='black', linewidth=3)
    ax.plot([0.4, 0.4], [0.3, 0.4], color='black', linewidth=3)

    # Nose (triangle)
    ax.plot([0, -0.1], [0.1, 0], color='black')
    ax.plot([0, 0.1], [0.1, 0], color='black')
    ax.plot([-0.1, 0.1], [0, 0], color='black')

    # Mouth
    ax.plot([0, -0.1], [0, -0.1], color='black')
    ax.plot([0, 0.1], [0, -0.1], color='black')

    # Whiskers
    for y in [0, 0.05, -0.05]:
        ax.plot([-1.2, -0.2], [y, y], color='black')  # Left whiskers
        ax.plot([0.2, 1.2], [y, y], color='black')    # Right whiskers

    plt.xlim(-1.5, 1.5)
    plt.ylim(-1.2, 1.6)
    plt.show()

draw_cat_with_lines()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def draw_disco_cat():
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.axis('off')
    plt.xlim(-2, 2)
    plt.ylim(-2, 3)

    # Elements to animate
    body_lines = []
    arm_left = ax.plot([], [], color='black', linewidth=2)[0]
    arm_right = ax.plot([], [], color='black', linewidth=2)[0]
    leg_left = ax.plot([], [], color='black', linewidth=2)[0]
    leg_right = ax.plot([], [], color='black', linewidth=2)[0]

    # Static elements (draw once)
    def draw_static():
        # Head
        head = plt.Circle((0, 1.5), 0.4, fill=False, linewidth=2)
        ax.add_patch(head)

        # Ears
        ax.plot([-0.3, -0.4, -0.2], [1.8, 2.1, 2.0], color='black')  # Left ear
        ax.plot([0.3, 0.4, 0.2], [1.8, 2.1, 2.0], color='black')     # Right ear

        # Eyes
        ax.plot([-0.15, -0.15], [1.5, 1.55], color='black', linewidth=2)
        ax.plot([0.15, 0.15], [1.5, 1.55], color='black', linewidth=2)

        # Nose + Mouth
        ax.plot([0, -0.05], [1.45, 1.4], color='black')
        ax.plot([0, 0.05], [1.45, 1.4], color='black')
        ax.plot([-0.05, 0.05], [1.4, 1.4], color='black')

        # Whiskers
        for y in [1.4, 1.42, 1.38]:
            ax.plot([-0.3, -0.5], [y, y], color='black')
            ax.plot([0.3, 0.5], [y, y], color='black')

        # Body
        body_lines.append(ax.plot([0, 0], [1.1, 0.5], color='black', linewidth=2)[0])

        # Disco ball
        disco = plt.Circle((0, 2.5), 0.3, fill=False, linewidth=1.5)
        ax.add_patch(disco)

        # Disco sparkles
        for dx in [-0.3, 0, 0.3]:
            ax.plot([0+dx, 0+dx], [2.8, 2.7], color='gray', linestyle='--')
        for dy in [2.5]:
            ax.plot([-0.3, -0.4], [dy, dy + 0.2], color='gray', linestyle='--')
            ax.plot([0.3, 0.4], [dy, dy + 0.2], color='gray', linestyle='--')

    # Animation update function
    def update(frame):
        # Simple oscillation
        angle = (-1)**frame * 0.3

        # Arms
        arm_left.set_data([0, -0.5], [1.0, 1.0 + angle])
        arm_right.set_data([0, 0.5], [1.0, 1.0 - angle])

        # Legs
        leg_left.set_data([0, -0.3], [0.5, 0.1 + angle])
        leg_right.set_data([0, 0.3], [0.5, 0.1 - angle])

        return [arm_left, arm_right, leg_left, leg_right]

    draw_static()
    ani = animation.FuncAnimation(fig, update, frames=20, interval=300, blit=True)
    plt.show()

draw_disco_cat()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.patches as patches

def draw_disco_cat_with_plane_and_mountain():
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.axis('off')
    plt.xlim(-3, 3)
    plt.ylim(-2, 3.5)

    # Elements to animate
    arm_left = ax.plot([], [], color='black', linewidth=2)[0]
    arm_right = ax.plot([], [], color='black', linewidth=2)[0]
    leg_left = ax.plot([], [], color='black', linewidth=2)[0]
    leg_right = ax.plot([], [], color='black', linewidth=2)[0]

    def draw_static():
        # --- Background ---
        # Sky
        ax.set_facecolor('#001f3f')  # Deep blue

        # Mountains
        mountain = patches.Polygon(
            [[-3, -2], [-1.5, 1], [0, -2]], closed=True, color="#2e8b57")
        mountain2 = patches.Polygon(
            [[0, -2], [1.5, 1.2], [3, -2]], closed=True, color="#3cb371")
        ax.add_patch(mountain)
        ax.add_patch(mountain2)

        # Ground
        ax.add_patch(patches.Rectangle((-3, -2), 6, 0.5, color="#654321"))

        # --- Disco Ball ---
        disco = patches.Circle((0, 2.8), 0.3, color='silver', ec='white', linewidth=2)
        ax.add_patch(disco)

        # Disco sparkles
        for dx in [-0.3, 0, 0.3]:
            ax.plot([0+dx, 0+dx], [3.1, 3], color='yellow', linestyle='--', linewidth=1)
        ax.plot([-0.3, -0.5], [2.8, 3], color='magenta', linestyle='--')
        ax.plot([0.3, 0.5], [2.8, 3], color='cyan', linestyle='--')

        # --- Cat ---
        # Head
        head = patches.Circle((0, 1.5), 0.4, color='#f4c2c2', ec='black', linewidth=2)
        ax.add_patch(head)

        # Ears
        ax.plot([-0.3, -0.4, -0.2], [1.8, 2.1, 2.0], color='black')  # Left ear
        ax.plot([0.3, 0.4, 0.2], [1.8, 2.1, 2.0], color='black')     # Right ear

        # Eyes
        ax.plot([-0.15], [1.55], marker='o', color='green', markersize=6)
        ax.plot([0.15], [1.55], marker='o', color='green', markersize=6)

        # Nose
        ax.plot([0], [1.48], marker='^', color='pink', markersize=5)

        # Mouth
        ax.plot([0, -0.05], [1.45, 1.4], color='black')
        ax.plot([0, 0.05], [1.45, 1.4], color='black')

        # Whiskers
        for y in [1.4, 1.42, 1.38]:
            ax.plot([-0.3, -0.5], [y, y], color='black')
            ax.plot([0.3, 0.5], [y, y], color='black')

        # Body
        ax.plot([0, 0], [1.1, 0.5], color='black', linewidth=2)

        # --- Plane ---
        plane_body = ax.plot([-2.5, 2.5], [2.5, 2.5], color='darkblue', linewidth=2)[0]  # Plane body
        ax.plot([2.5, 3], [2.5, 2.7], color='darkblue', linewidth=2)  # Right wing
        ax.plot([-2.5, -3], [2.5, 2.7], color='darkblue', linewidth=2)  # Left wing

        ax.plot([0, 0], [2.5, 2.6], color='darkblue', linewidth=2)  # Plane tail

        # --- Plane Trail ---
        ax.plot([-1.5, -0.5], [2.5, 2.8], color='white', linestyle='--', linewidth=1)

    def update(frame):
        angle = (-1)**frame * 0.3

        # Arms
        arm_left.set_data([0, -0.5], [1.0, 1.0 + angle])
        arm_right.set_data([0, 0.5], [1.0, 1.0 - angle])

        # Legs
        leg_left.set_data([0, -0.3], [0.5, 0.1 + angle])
        leg_right.set_data([0, 0.3], [0.5, 0.1 - angle])

        return [arm_left, arm_right, leg_left, leg_right]

    draw_static()
    ani = animation.FuncAnimation(fig, update, frames=20, interval=300, blit=True)
    plt.show()

draw_disco_cat_with_plane_and_mountain()
```
